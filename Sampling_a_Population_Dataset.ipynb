```python
#importing libraries

import pandas as pd
import seaborn as sns
import numpy as np
import re

import matplotlib
import matplotlib.pyplot as plt
plt.style.use ('ggplot')
from matplotlib.pyplot import figure
```

```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import io
 
df = pd.read_csv(io.BytesIO(uploaded['population_total.csv']))
print(df)
```

```python
#checking the dataset
df
```

```python
df.dtypes
```

```python
#dropping Null values
df.dropna(inplace = True)
```

```python
#making a pivot table
pivot_populaton = df.pivot(index = "year", 
                        columns = "country", 
                        values = "population" )
```

```python
#selecting some countries
pivot_populaton = pivot_populaton[['Nigeria', 'Ghana', 'Niger','Cameroon','Senegal','Benin','Togo']]
```

```python
#Showing Pivot table
pivot_populaton
```

```python
pivot_populaton.dtypes
```

```python
#changing the type of columns
pivot_populaton['Nigeria'] = pivot_populaton['Nigeria'].astype('int64')
pivot_populaton['Ghana'] = pivot_populaton['Ghana'].astype('int64')
pivot_populaton['Niger'] = pivot_populaton['Niger'].astype('int64')
pivot_populaton['Cameroon'] = pivot_populaton['Cameroon'].astype('int64')
pivot_populaton['Senegal'] = pivot_populaton['Senegal'].astype('int64')
pivot_populaton['Benin'] = pivot_populaton['Benin'].astype('int64')
pivot_populaton['Togo'] = pivot_populaton['Togo'].astype('int64')
```

```python
pivot_populaton
```

```python
pivot_populaton.plot(kind = 'line', xlabel = 'Year', ylabel = 'Population', 
                     title = 'Population (1955 -2020)',
                    figsize = (14,6))
#save plot
plt.savefig('pnd_proj.png')
#show plot
plt.show()
```

```python
#plotting a barplot for year (2020)
#using the .index attribute and isin method
pivot_populaton_2020 = pivot_populaton[pivot_populaton.index.isin([2020])]
```

```python
#showing new dataframe
pivot_populaton_2020
```

```python
#transpose dataframe from rows to columns, when making barplots, text data has to be in the index
pivot_populaton_2020.T
```

```python
pivot_populaton_chrt = pivot_populaton_2020.T
```

```python
pivot_populaton_chrt
```

```python
# making barplot
pivot_populaton_chrt.plot(kind='bar',color= 'green',
                         xlabel = 'Year', ylabel = 'Population',
                         title = 'Population 2020',
                           figsize = (8,4))
#show plot
plt.show()
```

```python
#selecting few years
pivot_populaton_chrtgrpd = pivot_populaton[pivot_populaton.index.isin([2000,2005,2010,2015,2020])]
```

```python
pivot_populaton_chrtgrpd
```

```python
#making grouped barplots
pivot_populaton_chrtgrpd.plot(kind='bar',  xlabel = 'Year', ylabel = 'Population',
                         title = 'Population last two decades',
                           figsize = (14,6))
 #show plot
plt.show()                            
```

```python
#changing the Column name
#its not good practice to use integer as columns in piechart,so i will cahnge the name of the column to string
#using the .rename() method
pivot_populaton_chrt.rename(columns = {2020:'2020'}, inplace = True)
```

```python
colors = ["#2ca02c", "#ff7f0e", "#1f77b4", "#d62728", "#9467bd", "#8c564b", "#7f00ff" ]
#colors = ['red', 'gold', 'yellowgreen', 'blue', 'lightcoral', 'lightskyblue', 'violet']

      #"#2ca02c"  "#1f77b4"   
```

```python
explode = (0.1, 0, 0, 0, 0, 0,0)
```

```python
#piechart plot

pivot_populaton_chrt.plot(kind = 'pie', y = '2020',
                     radius = 2.0,
                     colors = colors,autopct='%1.1f%%', explode = explode, shadow=True, startangle=140)
plt.title('POPULATION(%)', pad = 120, bbox={'facecolor':'0.8', 'pad':5})
plt.legend(bbox_to_anchor=(1.50, 1), loc='upper left', borderaxespad=0)
             
 
 #show plot
plt.show()   
```
